```python
import pandas as pd

# Load the dataset
file_path = '/content/Africa_climate_change.csv'
data = pd.read_csv(file_path)

# Display the first few rows to understand the structure
data.head()
```

```python
# Filter data for Tunisia and Cameroon
filtered_data = data[data['COUNTRY'].isin(['Tunisia', 'Cameroon'])]
```

```python
# Group by year and country, then calculate the average temperature
avg_temp = filtered_data.groupby(['DATE', 'COUNTRY'])['TAVG'].mean().reset_index()
```

```python
# Plot the data

import pandas as pd
import matplotlib.pyplot as plt # Import the matplotlib.pyplot module

# Load the dataset
file_path = '/content/Africa_climate_change.csv'
data = pd.read_csv(file_path)
plt.figure(figsize=(12, 6))

for country in ['Tunisia', 'Cameroon']:
    country_data = avg_temp[avg_temp['COUNTRY'] == country]
    plt.plot(country_data['DATE'], country_data['TAVG'], label=country)

# Customize the plot
plt.title('TAVG Fluctuations in Tunisia and Cameroon', fontsize=14)
plt.xlabel('DATE', fontsize=12)
plt.ylabel('TAVG (°C)', fontsize=12)
plt.legend(title='COUNTRY')
plt.grid(True)

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt # Import the matplotlib.pyplot module

# Load the dataset
file_path = '/content/Africa_climate_change.csv'
data = pd.read_csv(file_path)

senegal_data = data[data['COUNTRY'] == 'Senegal']

# Convert 'DATE' column to numeric, errors='coerce' will handle non-numeric values
senegal_data['DATE'] = pd.to_numeric(senegal_data['DATE'], errors='coerce')

# Filter data for 1980-2000 and 2000-2023
senegal_1980_2000 = senegal_data[(senegal_data['DATE'] >= 1980) & (senegal_data['DATE'] <= 2000)]
senegal_2000_2023 = senegal_data[(senegal_data['DATE'] > 2000) & (senegal_data['DATE'] <= 2023)]

# Calculate average temperature for each period
senegal_1980_2000['TAVG'] = senegal_1980_2000['TAVG'].mean() # Calculate average temperature and add it as a new column
senegal_2000_2023['TAVG'] = senegal_2000_2023['TAVG'].mean() # Calculate average temperature and add it as a new column


# Create the figure
plt.figure(figsize=(12, 6))

# Histogram for 1980-2000
plt.hist(senegal_1980_2000['TAVG'], bins=15, alpha=0.5, label='1980-2000', color='blue')

# Histogram for 2000-2023
plt.hist(senegal_2000_2023['TAVG'], bins=15, alpha=0.5, label='2000-2023', color='orange')

# Customize the plot
plt.title('Temperature Distribution in Senegal (1980-2000 vs 2000-2023)', fontsize=14)
plt.xlabel('TAVG (°C)', fontsize=12)
plt.ylabel('Frequency', fontsize=12)
plt.legend(title='Period')
plt.grid(True)

# Show the plot
plt.show()
```

```python
# Select the best chart to show the Average temperature per country.

import matplotlib.pyplot as plt

# Group by country to calculate average temperature
avg_temp_country = data.groupby('COUNTRY')['TAVG'].mean().sort_values()

# Plot the bar chart
plt.figure(figsize=(12, 6))
avg_temp_country.plot(kind='bar', color='skyblue')

# Customize the chart
plt.title('TAVG per Country', fontsize=14)
plt.xlabel('Country', fontsize=12)
plt.ylabel('TAVG (°C)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```
