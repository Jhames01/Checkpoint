```python
#Use a while loop to create a menu of options for the user to add, remove, or view items from the list.
def display_menu():
    print("\nMenu:")
    print("1. Add item")
    print("2. Remove item")
    print("3. View items")
    print("4. Exit")

def add_item(item_list):
    item = input("Enter the item to add: ")
    item_list.append(item)
    print(f"Item '{item}' added.")

def remove_item(item_list):
    item = input("Enter the item to remove: ")
    if item in item_list:
        item_list.remove(item)
        print(f"Item '{item}' removed.")
    else:
        print(f"Item '{item}' not found in the list.")

def view_items(item_list):
    if item_list:
        print("Items in the list:")
        for i, item in enumerate(item_list, start=1):
            print(f"{i}. {item}")
    else:
        print("The list is empty.")

def main():
    item_list = []
    while True:
        display_menu()
        choice = input("Choose an option (1-4): ")

        if choice == '1':
            add_item(item_list)
        elif choice == '2':
            remove_item(item_list)
        elif choice == '3':
            view_items(item_list)
        elif choice == '4':
            print("Exiting the program.")
            break
        else:
            print("Invalid choice. Please enter a number between 1 and 4.")

if __name__ == "__main__":
    main()
```

```python
#Use the list, tuple, set, and dictionary data structures to store and manipulate the shopping items.
def display_menu():
    print("\nMenu:")
    print("1. Add item to the list (up to a maximum of 5 items)")
    print("2. Remove item from the list")
    print("3. View items in the list")
    print("4. Add item with quantity")
    print("5. View item quantities")
    print("6. Add item to category (unique items only)")
    print("7. View items in category")
    print("8. Exit")

def add_item_to_list(item_list, max_items):
    if len(item_list) >= max_items:
        print(f"Cannot add more items. The list is limited to {max_items} items.")
        return

    item = input("Enter the item to add: ")
    item_list.append(item)
    print(f"Item '{item}' added to the list.")

def remove_item_from_list(item_list):
    item = input("Enter the item to remove: ")
    if item in item_list:
        item_list.remove(item)
        print(f"Item '{item}' removed from the list.")
    else:
        print(f"Item '{item}' not found in the list.")

def view_items_in_list(item_list):
    if item_list:
        print("Items in the list:")
        for i, item in enumerate(item_list, start=1):
            print(f"{i}. {item}")
    else:
        print("The list is empty.")

def add_item_with_quantity(item_dict):
    item = input("Enter the item name: ")
    quantity = int(input("Enter the quantity: "))
    item_dict[item] = item_dict.get(item, 0) + quantity
    print(f"Item '{item}' added with quantity {quantity}.")

def view_item_quantities(item_dict):
    if item_dict:
        print("Items and quantities:")
        for item, quantity in item_dict.items():
            print(f"{item}: {quantity}")
    else:
        print("No items in the dictionary.")

def add_item_to_category(item_set, category):
    item = input("Enter the item to add to category: ")
    if item in item_set:
        print(f"Item '{item}' is already in the category.")
    else:
        item_set.add(item)
        print(f"Item '{item}' added to the category.")

def view_items_in_category(item_set):
    if item_set:
        print("Items in the category:")
        for item in item_set:
            print(item)
    else:
        print("The category is empty.")

def main():
    item_list = []
    max_items = 5

    category_items = set()
    item_dict = {}

    categories = ("Fruits", "Vegetables", "Dairy")

    while True:
        display_menu()
        choice = input("Choose an option (1-8): ")

        if choice == '1':
            add_item_to_list(item_list, max_items)
        elif choice == '2':
            remove_item_from_list(item_list)
        elif choice == '3':
            view_items_in_list(item_list)
        elif choice == '4':
            add_item_with_quantity(item_dict)
        elif choice == '5':
            view_item_quantities(item_dict)
        elif choice == '6':
            print("Available categories:")
            for i, category in enumerate(categories, start=1):
                print(f"{i}. {category}")
            category_choice = int(input("Choose a category number: "))
            if 1 <= category_choice <= len(categories):
                category = categories[category_choice - 1]
                print(f"Category selected: {category}")
                add_item_to_category(category_items, category)
            else:
                print("Invalid category number.")
        elif choice == '7':
            print("Items in the current category:")
            view_items_in_category(category_items)
        elif choice == '8':
            print("Exiting the program.")
            break
        else:
            print("Invalid choice. Please enter a number between 1 and 8.")

if __name__ == "__main__":
    main()
```
